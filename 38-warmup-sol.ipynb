The following picture shows the network of Florentine families in early 15th century

![Florentine families](https://datarepository.wolframcloud.com/resources/images/3ea/3ea30ebd-fcac-4105-b886-b6ddd63dcbd3-io-1-o.en.gif)

A node represents a family. An edge means there is a marriage relationship between the two families

- Use networkx to represent the network. Then re-draw the network
- What is the largest clique in the network

```python
import networkx as nx
%matplotlib inline
```

```python
net = nx.Graph()
```

```python
net.add_node("Acciaiuoli")
net.add_node("Medici")
net.add_node("Castellani")
net.add_node("Peruzzi")
net.add_node("Strozzi")
net.add_node("Barbadori")
net.add_node("Ridolfi")
net.add_node("Tornabuoni")
net.add_node("Albizzi")
net.add_node("Salviati")
net.add_node("Pazzi")
net.add_node("Bischeri")
net.add_node("Guadagni")
net.add_node("Ginori")
net.add_node("Lamberteschi")
```

```python
net.add_edge("Acciaiuoli", "Medici")
net.add_edge("Castellani", "Peruzzi")
net.add_edge("Castellani", "Strozzi")
net.add_edge("Castellani", "Barbadori")
net.add_edge("Medici", "Barbadori")
net.add_edge("Medici", "Ridolfi")
net.add_edge("Medici", "Tornabuoni")
net.add_edge("Medici", "Albizzi")
net.add_edge("Medici", "Salviati")
net.add_edge("Salviati", "Pazzi")
net.add_edge("Peruzzi", "Strozzi")
net.add_edge("Peruzzi", "Bischeri")
net.add_edge("Strozzi", "Ridolfi")
net.add_edge("Strozzi", "Bischeri")
net.add_edge("Ridolfi", "Tornabuoni")
net.add_edge("Tornabuoni", "Guadagni")
net.add_edge("Albizzi", "Ginori")
net.add_edge("Albizzi", "Guadagni")
net.add_edge("Bischeri", "Guadagni")
net.add_edge("Guadagni", "Lamberteschi")
```

```python
nx.draw(net, with_labels = True)
```

```python
cliques = [c for lc in nx.cliques_containing_node(net).values()
 for c in lc ]
```

```python
cliques = sorted(cliques, key=len, reverse=True )
```

```python
largest_size = len(cliques[0])
[c for c in cliques if len(c) == largest_size]
```

```python
nx.write_graphml(net, "data/florentine.graphml")
```
