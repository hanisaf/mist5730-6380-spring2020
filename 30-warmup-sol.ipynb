```python
import pandas as pd
import altair as alt
alt.renderers.enable('notebook')
```

0- Load the Gapminder dataset

```python
data = pd.read_csv('data/gapminder.tsv', sep='\t')
```

```python
data.head()
```

1- Relate per capita GDP to life expectancy using point mark

```python
alt.Chart(data).mark_point().encode(
    x='gdpPercap',
    y='lifeExp'
)
```

2- Use color to differentiate between the different years

```python
alt.Chart(data).mark_point().encode(
    x='gdpPercap',
    y='lifeExp',
    color='year',
    tooltip='country'
)
```

4- Can you account for population size? Try changing the size of the dots

```python
alt.Chart(data).mark_point().encode(
    x='gdpPercap',
    y='lifeExp',
    color='year',
    tooltip='country',
    size='pop'
)
```

```python
alt.Chart(data).mark_point().encode(
    x='gdpPercap',
    y='lifeExp',
    color='year',
    tooltip='country',
    size=alt.Size('pop', scale = alt.Scale(range=(50,1500)))
)
```

5- Use facets (cols/rows) to break this plot into one plot per year

```python
alt.Chart(data).mark_point().encode(
    x='gdpPercap',
    y='lifeExp',
    row='year',
    tooltip='country',
    size='pop'
)
```

6- how does life expectancy vary per continent?

```python
alt.Chart(data).mark_point().encode(
    x='continent',
    y='average(lifeExp)',
    color='year'
)
```

```python
alt.Chart(data).mark_bar().encode(
    y='continent',
    x='average(lifeExp)',
    #color='year'
)
```

```python
chart1 = alt.Chart(data).mark_bar().encode(
    y='continent',
    x='min(lifeExp)',
    x2='max(lifeExp)',
    #color='year'
)
```

```python
chart2 = alt.Chart(data).mark_tick(color='red').encode(
    y='continent',
    x='average(lifeExp)',
    #color='red'
)
```

```python
chart1 + chart2
```

7- how did life expectancy change over time?

```python
alt.Chart(data).mark_line().encode(
    x='year',
    y='average(lifeExp)',
    color='continent'
)
```

```python
alt.Chart(data).mark_line().encode(
    x='year',
    y='average(lifeExp)',
    row='continent',
    color='country',
    tooltip='country'
)
```
